```python
# https://likegeeks.com/python-dijkstras-algorithm/
from numpy import inf
```

```python
# Adjacency List Representation
graph = {'A': {'C': 5, 'D': 1, 'E': 2}, 'B': {'H': 1, 'G': 3}, 'C': {'I': 2, 'D': 3, 'A': 5},
         'D': {'C': 3, 'A': 1, 'H': 2}, 'E': {'A': 2, 'F': 3},
         'F': {'E': 3, 'G': 1}, 'G': {'F': 1, 'B': 3, 'H': 2}, 'H': {'I': 2, 'D': 2, 'B': 1, 'G': 2},
         'I': {'C': 2, 'H': 2}}
```

![](01-Example-Graph.jpg)

```python
costs = {'A': 0, 'B': inf, 'C': inf, 'D': inf, 'E': inf, 'F': inf, 'G': inf, 'H': inf, 'I': inf}
parents = {}

## Adjacency Matrix Representation
#import numpy as np
#n=9
#adjacency_matrix_graph=np.zeros((n,n))
```

```python
def search(source, target, graph, costs, parents):
    nextNode = source  # A is source
    
    while nextNode != target: # while A is not target
        
        for neighbor in graph[nextNode]: # Neighbour(s) of A = {C, D, E}
            
            if graph[nextNode][neighbor] + costs[nextNode] < costs[neighbor]: 
                
                costs[neighbor] = graph[nextNode][neighbor] + costs[nextNode]
                
                parents[neighbor] = nextNode
                
            del graph[neighbor][nextNode] # Delete node A from evaluation set
            
        del costs[nextNode]
        
        nextNode = min(costs, key=costs.get) # Find new node for evaluation from the minimum costs
        
    return parents
```

```python
result = search('A', 'G', graph, costs, parents)
```

```python
def backpedal(source, target, searchResult):
    
    node = target
    
    backpath = [target]
    
    path = []
    
    while node != source:
        
        backpath.append(searchResult[node])
        
        node = searchResult[node]
        
    for i in range(len(backpath)):
        
        path.append(backpath[-i - 1])
        
    return path
```

```python
print('parent dictionary={}'.format(result))

print('shortest path={}'.format(backpedal('A', 'G', result)))
```
