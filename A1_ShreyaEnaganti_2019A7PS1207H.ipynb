```python
"""
Develop a circle and line representation of a robot.Take user input of a position (x,y) and
orientation (theta) to place the robot in the plot.
"""

import numpy as np
import matplotlib.pyplot as plt
x = int(input("X"))
y = int(input("Y"))
theta = float(input("Theta"))

plt.plot(x, y, 'ro', markersize=11)  
plt.plot(x+0.2*np.cos(theta), y+0.2*np.sin(theta), 'bo')
plt.plot(x+0.3*np.cos(theta), y+0.3*np.sin(theta), 'bo')
plt.plot(x+0.4*np.cos(theta), y+0.4*np.sin(theta), 'bo')
plt.plot(x+0.5*np.cos(theta), y+0.5*np.sin(theta), 'bo')
plt.plot(x+0.6*np.cos(theta), y+0.6*np.sin(theta), 'bo')
plt.axis([0, 10, 0, 10])              
plt.show()
```

```python
"""
Generate a parametric equation for a trajecctory (x vs t and y vs t, theta vs t, 0<=t<=1),
where the starting point is at the origin of the Global frame, and plot it.
"""

import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 1, 0.001)
 
theta = (2 * np.pi * t)
radius = 0.1
x = radius * np.cos(theta)
y = radius * np.sin(theta)
 
figure, axes = plt.subplots( 1 )
 
axes.plot( x, y )
axes.set_aspect( 1 )
 
plt.title( 'Parametric Equation' )
plt.show()
```

```python
"""
Also find the
linear and angular velocities by diffierntiating the parametric equations and plot.
"""

linear_velocity = (-1) * radius * np.sin(theta) * 2 * np.pi
angular_velocity = linear_velocity / radius

print("Linear Velocity: ", linear_velocity)
print("Angular Velocity: ", angular_velocity)
```

```python
"""
Divide the path parametrically in many small segments, use a loop to plot the agent at
those posees generated by the parametric equations. (orientation of the robot can be
found from the slope of the trajectory).
"""

t1 = np.arange(0, 1, 0.01)
theta1 = (2 * np.pi * t1)

for theta_t in theta1:
    t = np.arange(0, 1, 0.001)
    theta = (2 * np.pi * t)
    radius = 0.1
    x = radius * np.cos(theta)
    y = radius * np.sin(theta)
    figure, axes = plt.subplots( 1 )
    axes.plot( x, y )
    axes.set_aspect( 1 )
    #plt.show()
    x1 = radius * np.cos(theta_t)
    y1 = radius * np.sin(theta_t)
    plt.plot(x1, y1, 'ro', markersize=11) 
    plt.plot(radius*np.cos(theta_t+0.05), radius*np.sin(theta_t+0.05), 'bo')
    plt.plot(radius*np.cos(theta_t+0.07), radius*np.sin(theta_t+0.07), 'bo')
    plt.plot(radius*np.cos(theta_t+0.09), radius*np.sin(theta_t+0.09), 'bo')
    plt.plot(radius*np.cos(theta_t+0.11), radius*np.sin(theta_t+0.11), 'bo')
    plt.show()
    #figure.canvas.draw()
    plt.close()
    
#from moviepy.video.io.bindings import mplfig_to_npimage
#import matplotlib.pyplot as plt
#numpy_fig = mplfig_to_npimage(figure)
    
#make video
```

```python
"""
Develop a video based on the series of plots generated for the agent. (Take at least 100
parametric steps such that if you are going to make 24 fps video it will generate ~4 s
video file)
"""

# importing required libraries
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.animation as animation
from IPython import display

# initializing a figure
fig = plt.figure()

# labeling the x-axis and y-axis
axis = plt.axes(xlim=(-0.25, 0.25), ylim=(-0.25, 0.25))

# initializing a line variable
line1, = axis.plot([], [], lw=3)
line2, = axis.plot([], [], 'ro', markersize=1)

def animate(frame_number):
    t = np.arange(0, 1, 0.008)
    theta = (2 * np.pi * (t-0.01*frame_number))
    radius = 0.2
    x = radius * np.cos(theta)
    y = radius * np.sin(theta)
    line1.set_data(x, y)
    line1.set_color('green')
    x1 = []
    y1 = []
    for theta_t in theta:
        x1.append(radius * np.cos(theta_t))
        y1.append(radius * np.sin(theta_t))
    x2 = np.array(x1) 
    y2 = np.array(y1)
    line2.set_data(x2, y2)
    line2.set_color('blue')
    return line1, line2


anim = animation.FuncAnimation(fig, animate, frames=100, interval=20, blit=True)
fig.suptitle('Parametric Plot', fontsize=14)

video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```
